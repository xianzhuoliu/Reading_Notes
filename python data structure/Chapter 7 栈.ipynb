## 示例程序：匹配括号
检查步骤：
1. 扫描整个表达式，将开始的括号压入到栈中
2. 当遇到一个结束括号时，如果栈为空，或者如果栈顶的项不是相同类型的括号，括号不匹配
3. 如果它是正确类型的括号，从栈顶弹出一项，继续扫描表达式
4. 当到达了表达式的末尾，栈应该为空，如果栈不为空，括号不匹配。

````python
from stack import LinkedStack

def bracketsBalance(exp):          
    """exp represents the expression"""
    stk = LinkedStack()                      # Create a new stack
    for ch in exp:                   # Scan across the expression
        if ch in ['[', '(']:            # Push an opening bracket 
            stk.push(ch)
        elif ch in [']', ')']:        # Process a closing bracket
            if stk.isEmpty():                      # Not balanced
                return False                  
            chFromStack = stk.pop()
            # Brackets must be of same type and match up
            if ch == ']' and chFromStack != '[' or \
               ch == ')' and chFromStack != '(':
                return False
    return stk.isEmpty()                   # They all matched up


def main():
    exp = raw_input("Enter a bracketed expression: ")
    if bracketsBalance(exp):
        print "OK"
    else:
        print "Not OK"

main()
````

三个步骤:

在这个算法中，**标记token**指的是一个运算数或一个运算符



## 将中缀表达式转换为后缀表达式
大体上，这个算法从左向右扫描包含了中缀表达式的序列，并同时构建出包含了等价的后缀表达式的序列。
- 只要一遇到运算数，就会将其从中缀序列复制到后缀序列中
- 然而，先不要将运算符压入到栈中，直到在它之前所有具有较高优先级的运算符已经复制到了后缀字符串中，再将其压入到栈中。

详细步骤：
1. 开始时有一个空的后缀表达式和一个空的栈，栈用来保存运算符和左圆括号
2. 扫描中缀表达式
3. 遇到一个运算数的时候，将其添加到后缀表达式后面
4. 遇到左圆括号时，将其压入到栈中
5. 遇到一个运算符，从栈中弹出和它具有相等或更高优先级的所有运算符，将它们添加到后缀表达式的末尾，然后将扫描到的运算符压入到栈中
6. 遇到一个右圆括号的时候，将运算符从栈中移动到后缀表达式中，直到遇到左圆括号并将其丢弃
7. 中缀表达式结束，剩下运算符都转移到后缀表达式中



## 回溯算法
回溯算法从一个预定义的起始状态开始随后从一个状态移动到另一个状态。在这个过程中的任何时刻，当在多个可选状态的时候，算法就会选取一个状态。如果算法到达了表示不希望结果的一个状态，它会回到上一个拥有一个未探索的、可替代选项的位置。算法要么穷尽了对所有状态的搜索，要么达到了想要的最终状态。

主要有两种技术来实现回溯算法，一种是使用栈，还有一种是使用递归。

```
create an empty stack
push the starting state onto the stack
while the stack is not empty
    pop the stack and examine the state
    if the state represents an ending state
        return SUCESSFUL CONCLUSION
    Else if the state has not been visited previously
        mark the state as visited
        push onto the stack all unvisited adjacent states
return UNSUCCESSFUL CONCLUSION
```

## 将栈添加到集合层级中
包的三个实现LinkedBag, ArrayBag和ArraySortedBag都是两个抽象类AbstractBag和Abstract-Collection的子孙。

两个栈实现ArrayStack和LinkedStack也类似，它们从AbstractStack继承了add方法，从AbstractCollection继承了$_size变量，isEmpty, __len__, __str__,__add__,__eq__方法$，要实现的方法只有$__init__, peek, push, pop, clear, __iter__方法$

## 数组实现
pop方法其实最后一位还在内存里面，只是len(self)没有到它那里了，所有相当于被忘却了

````python
from arrays import Array
from abstractstack import AbstractStack

class ArrayStack(AbstractStack):
    """An array-based stack implementation."""

    # Class variable
    DEFAULT_CAPACITY = 10

    # Constructor
    def __init__(self, sourceCollection = None):
        """Sets the initial state of self, which includes the
        contents of sourceCollection, if it's present."""
        self._items = Array(ArrayStack.DEFAULT_CAPACITY)
        AbstractStack.__init__(self, sourceCollection)

    # Accessor methods
    def __iter__(self):
        """Supports iteration over a view of self.
        Visits items from bottom to top of stack."""
        cursor = 0
        while cursor < len(self):
            yield self._items[cursor]
            cursor += 1

    def peek(self):
        """Returns the item at the top of the stack.
        Precondition: the stack is not empty.
        Raises: KeyError if stack is empty."""
        if self.isEmpty():
            raise KeyError("The stack is empty")
        return self._items[len(self) - 1]

    # Mutator methods
    def clear(self):
        """Makes self become empty."""
        self._size = 0
        self._items = Array(ArrayStack.DEFAULT_CAPACITY)

    def push(self, item):
        """Inserts item at top of the stack."""
        # Resize array here if necessary
        self._items[len(self)] = item
        self._size += 1

    def pop(self):
        """Removes and returns the item at the top of the stack.
        Precondition: the stack is not empty.
        Raises: KeyError if stack is empty.
        Postcondition: the top item is removed from the stack."""
        if self.isEmpty():
            raise KeyError("The stack is empty")
        oldItem = self._items[len(self) - 1]
        self._size -= 1
        # Resize the array here if necessary
        return oldItem
````

## AbstractStack类的作用
应该给位于具体栈类和AbstractCollection之间的AbstractStack留下些什么

前面列出的栈接口没有add方法，尽管已经包含了push方法，它所做的事情和add方法相同，但还是有很多客户，**包括集合框架自身之中的重要客户都喜欢用add方法。**

为了解决这个问题并保持和其他集合接口的一致性，需要添加一个add方法，并且所有的栈类型都可以使用，合理的放置位置就是AbstractStack类

````python
from abstractcollection import AbstractCollection

class AbstractStack(AbstractCollection):
    """An abstract stack implementation."""

    # Constructor
    def __init__(self, sourceCollection = None):
        """Sets the initial state of self, which includes the
        contents of sourceCollection, if it's present."""
        AbstractCollection.__init__(self, sourceCollection)

    # Mutator methods
    def add(self, item):
        """Adds item to self."""
        self.push(item)
````

## 两种实现的时间和空间分析

除了$__iter__$方法，所有的栈方法都很简单，并且运行时间不大于$O(1)$

iter在链表实现中，使用了递归函数系统调用栈，内存线性增长。

空间需求方面：
- 链表栈n个节点，每个节点包含两个引用，一个引用项另一个引用用于next节点。此外还有一个变量指向栈顶节点，还有一个变量用于大小。总空间需求为2n+2。 空间效率$\frac{n}{2n+2}=>\frac{1}{2}$
- 数组栈，空间需求是规定的。n+2。当数组的装载因子大于0.5的时候，数组实现比链表具有更高的空间效率
