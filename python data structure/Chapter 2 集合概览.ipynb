# 集合的实现

集合也成为抽象数据类型(abstract data types, ADT)，ADT的用户只关心其接口

抽象用于隐藏或忽略不重要的细节。函数和方法是抽象的最小单位，类是下一个较大的单位，而模块是最大的单位。

# Summary
- 集合是保存0个或多个其他对象的对象
- 集合的5个主要类别：线性集合，层次集合，图集合，无序集合和有序集合
- 线性集合每个项都有唯一的前驱（除了第一项）和唯一后驱（除了最后一项）
- 图中的项有0个或多个后继或前驱
- 集合是可迭代的，可以用for来访问每一个项
- 抽象的数据类型是一组对象，以及这些对象上的操作。因此，集合被称为抽象数据类型
- 数据结构是表示集合中包含的数据的一个对象 A data structure is an object that represents the data contained in a collection
