设计良好的软件一个特征就是，接口（interface）和实现（implementation）可以清晰地区分开来。当程序员使用一个软件资源的时候，他们只需要与其接口相连接。**这个接口是一组方法、函数、以及这些方法和函数可用的数据类型的名称。**理想情况下，这些资源如何实现的细节，也就是底层算法和数据结构，都封装在一个边界之中。**这个边界将接口和实现隔开。**

## 完成迭代器
之前Aaary类中的__iter__方法是在底层列表对象上调用iter函数，但实际上数组可能并不是填满的，但iter函数总会访问所有位置。为了解决这个问题。

yield语句可以将每一项都发给for的调用

```python
def __iter__(self):
    """Supports iteration over a view of self."""
    cursor = 0
    while cursor < len(self):
        yield self._items[cursor]
        cursor += 1
```

## 完成使用迭代器的方法
$__add__$方法将两个集合连接起来。$__str__$构建字符串

**注意$__add__$方法创建了self的副本，相当于把self作为soureCollection再生成一个具有同样元素的类对象。**

```python
def __str__(self):
    """Returns the string representation of self."""
    return "{" + ", ".join(map(str, self)) + "}"

def __add__(self, other):
    """Returns a new bag containing the contents
    of self and other."""
    result = ArrayBag(self)
    for item in other:
        result.add(item)
    return result

def __eq__(self, other):
    """Returns True if self equals other,
    or False otherwise."""
    if self is other: return True
    if type(self) != type(other) or \
       len(self) != len(other):
        return False
    for item in self:
        if not item in other:
            return False
    return True
```

## 初始化数据结构
self._items现在是外部指针而不是数组，改为None。导入Node类型。其他几乎是相同的。

```python
from node import Node

class LinkedBag(object):
    """A link-based bag implementation."""

    # Constructor
    def __init__(self, sourceCollection = None):
        """Sets the initial state of self, which includes the
        contents of sourceCollection, if it's present."""
        self._items = None
        self._size = 0
        if sourceCollection:
            for item in sourceCollection:
                self.add(item)
```

## 完成remove方法
我们想从ArrayBag的实现中尽可能多地借用代码，只有当需要的嘶吼，才在链表结构中操作指针。检查先验条件和搜索循环都和之前相同，现在的区别在于，我们在搜索中记录了两个指针，probe和trailer

```python
def remove(self, item):
        """Precondition: item is in self.
        Raises: KeyError if item in not in self.
        Postcondition: item is removed from self."""
        # Check precondition and raise if necessary
        if not item in self:
            raise KeyError(str(item) + " not in bag")
        # Search for the node containing the target item
        # probe will point to the target node, and trailer
        # will point to the one before it, if it exists
        probe = self._items
        trailer = None
        for targetItem in self:
            if targetItem == item:
                break
            trailer = probe
            probe = probe.next
        # Unhook the node to be deleted, either the first one or one
        # thereafter
        if probe == self._items:
            self._items = self._items.next
        else:
            trailer.next = probe.next
        # Decrement logical size
        self._size -= 1
```

# Summary
- 接口是软件资源用户可用的一组操作
- 接口中的内容是函数头、方法头，以及它们的文档
- 实现是必须符合接口的函数、方法和类
- 可以根据接口来指定集合类型，一个集合类型拥有几个实现类
- 多态是在两个或多个实现中使用相同的运算符号、函数名和方法。多态函数的示例是str和len，多态运算符的示例是+和==。多态方法的示例是add和isEmpty
- 包集合类型是无序的，并且支持添加项、删除项、和访问项等操作
- 类图是描述类之间关系的一种可视化方法
