```python
class Node(object):
    def __init__(self, data, next = None):
        self.data = data
        self.next = next
    def __str__(self):
        probe = self
        temp = []
        while probe != None:
            temp.append(probe.data)
            probe = probe.next
        return str(temp)
        
class TwoWayNode(Node):
    def __init__(self, data, previous = None, next = None):
        Node.__init__(self, data, next)
        self.previous = previous
        
def length(Node):
    sum = 0
    probe = Node
    while probe != None:
        probe = probe.next
        sum += 1
    return sum

def insert(index, data, node):
    """insert a value and create a copy 暂时不能对node本身进行修改（不是定义在node类中的方法），只是增加了一个node然后返回一个新的开头指针"""
    probe = node
    if node is None or index <= 0:
        node = Node(data, probe)
        return node
    else:
        while index > 1 and probe.next != None:
            probe = probe.next
            index -= 1
        probe.next = Node(data, probe.next)
        return node

def pop(index, head):
    if index <= 0 or head.next is None:
        removedItem = head.data
        head = head.next
        return head, removedItem
    else:
        probe = head
        while index > 1 and probe.next.next != None:
            probe = probe.next
            index -= 1
        removedItem = probe.next.data
        probe.next = probe.next.next
        return head, removedItem

def makeTwoWay(head):
    probe = head
    temp = []
    while probe != None:
        temp.append(probe.data)
        probe = probe.next
        
    tail = TwoWayNode(temp[0])
    for data in temp[1:]:
        tail.next = TwoWayNode(data, tail)
        tail = tail.next
    return tail
    
def main():
    head = None
    for i in range(1, 6):
        head = Node(i, head)
    print(head)
    
    len = length(head)
    print('the length of link: ', len)
    
    head = insert(0, 6, head)
    print(head)
    
    head, Item = pop(2, head)
    print(head)
    print(Item)
    
    tail = makeTwoWay(head)
    print(isinstance(tail, TwoWayNode))
    probe = tail
    while probe != None:
        print(probe.data)
        probe = probe.previous
    
if __name__ == '__main__':
    main()
```
