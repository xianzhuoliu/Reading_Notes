```python
"""A simple software system for a library models a library as a collection of books
and patrons. A patron can have at most three books out on loan at any given time.
A book also has a list of patrons waiting to borrow it. 
Each book has a title, an author, a patron to whom it has been checked out, and a list of patrons waiting for
that book to be returned. 
Each patron has a name and the number of books it has
currently checked out. 
Develop the classes Book and Patron to model these objects.
Think first of the interface or set of methods used with each class, and then choose
appropriate data structures for the state of the objects. Also, write a short script to
test these classes."""
```

```python
class Book(object):
    def __init__(self, title, author, browse = '', waitingList = ''):
        self.title = title
        self.author = author
        self.browse = browse
        self.waitingList = waitingList
    def info(self):
        if self.title == '' and self.author == '' and self.browse == '' and self.waitingList == '':
            print('this book doesn\'t exist')
        else:
            return self.title, self.author, self.browse, self.waitingList

class Patron(object):
    def __init__(self, name, borrow):
        self.name = name
        self.borrow = borrow
    def info(self):
        return self.name, self.borrow
    def action(self, book, flag):
        if flag == 0 and book.browse == '':               #0表示借
            if len(self.borrow) == 3:
                print('you at most borrow three books')
                return
            else:
                book.browse = self.name
                book.waitingList.remove(self.name)
                self.borrow.append(book.title)
                return
        if flag == 1:
            self.borrow.remove(book.title)
            book.browse = ''
            return
        if book.browse != '':
            book.waitingList.append(self.name)
            print('the book has been borrowed')
```

```python
book1 = Book('book1', 'A1', 'B1', ['B2','B3','B4'])
print(book1.info())
book2 = Book('book2', 'A2', 'B2', ['B1','B3','B4'])
book3 = Book('book3', 'A2', 'B2', ['B1','B3','B4'])
patron1 = Patron('B1',['book1'])
print(patron1.info())
patron1.action(book1,1)
print(patron1.info())
patron2 = Patron('B2',['book3','book2','book5'])
print(patron2.info())
patron2.action(book1,0)
patron2.action(book3, 1)
print(patron2.info())
patron2.action(book1, 0)
print(patron2.info())
patron1.action(book1,0)
print(book1.info())
```

```python
class Library(object):
    def addBook(self, title, author, browse = '', waitingList = ''):
        book = Book(title, author, browse, waitingList)
        return book.info()
    def delBook(self, book):
        book.title  = ''
        book.author = ''
        book.browse = ''
        book.waitingList = ''
    def findBook(self, book):
        return book.info()
    def borrowBook(self, book, browse):
        if book.browse == '':
            book.browse = browse
            if browse in book.waitingList :
                book.waitingList.remove(browse)
        else:
            print('the book has been borrowed')
    def returnBook(self, book):
            book.browse = ''
```

```python
admin = Library()
admin.addBook('book4', 'A3')
admin.delBook(book1)
admin.findBook(book1)
admin.borrowBook(book2, 'B2')
admin.returnBook(book2)
admin.borrowBook(book2, 'B2')
admin.findBook(book2)
```
